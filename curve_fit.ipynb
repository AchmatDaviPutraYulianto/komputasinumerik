```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
rng = np.random.RandomState(1)
x = 10 * rng.rand(50)
y = 2 * x - 5 + rng.randn(50)
plt.scatter(x, y);
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression(fit_intercept=True)

model.fit(x[:, np.newaxis], y)

xfit = np.linspace(0, 10, 1000)
yfit = model.predict(xfit[:, np.newaxis])

plt.scatter(x, y)
plt.plot(xfit, yfit);
```

```python
print("Model slope:    ", model.coef_[0])
print("Model intercept:", model.intercept_)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Inisialisasi generator angka acak
rng = np.random.RandomState(1)

# Data asli
x = 10 * rng.rand(50)
y = np.sin(x) + 0.1 * rng.randn(50)

# Membuat scatter plot
plt.figure(figsize=(8, 5))
plt.scatter(x, y, color='b', alpha=0.7, edgecolors='k', label='Data')
plt.plot(np.sort(x), np.sin(np.sort(x)), color='r', linestyle='-', linewidth=2, label='Garis Sinus')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Scatter Plot Data')
plt.legend()
plt.show()
```
